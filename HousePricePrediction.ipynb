```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
data =  fetch_california_housing()
df =pd.DataFrame(data.data,columns=data.feature_names)
df.head()
import matplotlib.pyplot as plt
```

```python
feature = df.iloc[:,:-1]
prediction =df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(feature,prediction,test_size=0.5,random_state=50,shuffle=True)
```

```python
Train_data =pd.concat([x_train,y_train],axis='columns',names=['MedInc','HouseAge','AveRooms',
'AveBedrms',
'Population',
'AveOccup',
'Latitude',
'Longitude',])
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
y_pred_train = model.predict(x_train)
y_pred =model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
mse_train =  mean_squared_error(y_train,y_pred_train)
mae_train = mean_absolute_error(y_train,y_pred_train)
r2_train = r2_score(y_train,y_pred_train)


print("Evolutioin metrics: Train Data")
print("mean Absolute Error",mae_train)
print("mean_squared error",mse_train)
print("R square Score ",r2_train)
```

```python

mse =  mean_squared_error(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)


print("Evolutioin metrics: Test Data")
print("mean Absolute Error",mae)
print("mean_squared error",mse)
print("R square Score ",r2)
```
